  A loop is a program construct that repeatedly executes the loop's statements (known as the loop body) while the loop's expression is true; when false, execution proceeds past the loop. Each time through a loop's statements is called an `iteration`.     
  In Python, we mainly have three types of loops:    
  for, while, and do-while.

#  while Loop Statements

Let’s look at an if statement and a while loop that use the same condition and take the same actions based on that condition. Here is the code with an if statement:

```python
count = 1

if count <= 5:
    print("Hello World")
```

Here is the code with a while statement:

```python
#------------------------------------- Method 1  "Condition-Controlled Loop"--------------------
count = 1
while count <= 5:  # Loop expression
    print("Hello")
    count = count + 1 # 6
```

```python
#------------------------------------- Method 2 "Break-Controlled Loop" -------------------------------
count = 1
while True:
    print("Hello World")
    count = count + 1
    if count > 5:
        break
```

# Deciding Which Loop to Use


**The while loop is a conditional pretest loop**   
*   Iterates as long as a certain condition exits  
*	Validating input   
*	Reading lists of data terminated by a sentinel

**The for loop is a pretest loop**
*   Built-in expressions for initializing, testing, and updating
*   Situations where the exact number of iterations is known
