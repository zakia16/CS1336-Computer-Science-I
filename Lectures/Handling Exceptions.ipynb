#  Raising exceptions

In this program, the user is asked to enter age and weight to estimate daily calories. The programmer adds error-checking using multiple if statements to ensure `age` and `weight` are within valid ranges.  
To ensure the user knows why a result wasn't produced, the programmer includes a final conditional check if `(age <= 0 or age > 120)` or `(weight <= 0)`. While effective, this creates redundancy, as the program has to evaluate the validity of the inputs twice by printing the "Cannot compute" message.

```python
#first program thatn doesn't use exception-handling constructs
user_input = ""

while user_input != "q":
    age = int(input("Enter age: "))

    if age <= 0 or age > 120:
        print("Invalid age.")
    else:
        weight = int(input("Enter weight (kg): "))

        if weight <= 0:
            print("Invalid weight.")
        else:
            calories = 10 * weight + 6.25 * 170 - 5 * age + 5
            print(f"Estimated calories: {calories}")

    if (age <= 0 or age > 120) or (weight <= 0):
        print("Cannot compute result.")

    user_input = input('Enter any key ("q" to quit): ')
```

Here is the code using exceptions:

```python
#second program that raises exception
user_input = ""

while user_input != "q":
    try:
        age = int(input("Enter age: "))
        if age <= 0 or age > 120:
            raise ValueError("Invalid age.")

        weight = int(input("Enter weight (kg): "))
        if weight <= 0:
            raise ValueError("Invalid weight.")

        calories = 10 * weight + 6.25 * 170 - 5 * age + 5
        print(f"Estimated calories: {calories}")

    except ValueError as excpt:
        print(excpt)
        print("Could not calculate result.")

    user_input = input('Enter any key ("q" to quit): ')
```

In the exception version, you only write the error handling once. If any error occurs at any point in the `try` block, it immediately jumps to the` except `block, preventing the calculation from ever running with bad data.      
Also, the first program has a hidden bug: if someone types "twenty" instead of 20, the `int()` function will throw an error and the program will crash.
