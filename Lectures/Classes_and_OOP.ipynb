A **class** is a blueprint for creating objects in Python.     
It is a programmer-defined type.
It defines the data (attributes) and behavior (methods) that the objects created from it will have.

When I say the word **car,** do you think of one specific car or cars in general?

•  That general idea is what we call a **class**.    
•  A specific car (your car, my car) is an **object**.

```python
class Car:
    def __init__(self):
        # Attributes that describe each car
        self.color = "unknown"
        self.speed = 0
        self.brand = "unknown"
        self.price = 1000

    def drive(self):
        # Behavior / action of the car
        print("Car is starting... vroom!")

# Creating first Car object
toyota = Car()
print(toyota.color, toyota.price)

# Creating 2nd Car object
ford = Car()
ford.color = "White"
ford.price = 20000
print(ford.color, ford.price)
```
