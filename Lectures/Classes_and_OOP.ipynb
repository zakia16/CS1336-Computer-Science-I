## break Statements


•	Stops the loop completely when a certain condition is met.   
•	The program immediately exits the loop.

```python
for num in range(1, 6):
    if num == 3:
        break  # Exits the loop when num is 3
    print(num)
```

##continue Statements

•	Skips the current iteration and moves to the next one.
•	The loop does not stop but continues with the next iteration.

```python
for num in range(1, 6):
    if num == 3:
        continue  # Skips the rest of the loop 			   # when num is 3
    print(num)
```

##While Loop using Sentinel value

A sentinel value is a special value used to signal the termination of a loop. In the example below, the loop continues until the user enters 'exit', which acts as the sentinel value to stop execution.

```python
user_input = input()

while user_input != 'exit':
	print(f"You entered: {user_input}")
	user_input = input("Type 'exit' to quit): ")

print("Exiting the program.")
```

#for Loops

A for loop iterates a specific number of times.     
General Format:

```python
for variable in container:
    # Loop body: Sub-statements to execute

# Statements to execute after the for loop is complete
```

Let see how to count using the range() function

```python
''' Print 0 to 5 using loop'''
for i in range(5):
    print(i, end=' ')
```

OR

```python
for i in range(0, 5):
    print(i, end=' ')
```

The range() function can also be called with three arguments. The first two arguments will be the start and stop values, and the third will be the step argument. The step is the amount by which the variable is increased after each iteration:

```python
for i in range(0, 5, 1):
    print(i, end=' ')
```

You can actually use a while loop to do the same thing as a for loop; for loops are just more concise.

```python
''' Print 0 to 5 using loop'''
count = 0
while count <= 5:
    print(count, end= ' ')
    count = count + 1
```


Reverse Counting using the range() function


```python
for i in range(4, -1, -1):
    print(i, end=' ')
```

# Deciding Which Loop to Use


**The while loop is a conditional pretest loop**   
*   Iterates as long as a certain condition exits  
*	Validating input   
*	Reading lists of data terminated by a sentinel

**The for loop is a pretest loop**
*   Built-in expressions for initializing, testing, and updating
*   Situations where the exact number of iterations is known
